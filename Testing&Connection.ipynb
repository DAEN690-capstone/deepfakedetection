```python
import os 
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.applications import InceptionResNetV2
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import InputLayer
from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import Model
from tensorflow.keras import optimizers
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.applications.inception_resnet_v2 import preprocess_input
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array, load_img
```

```python
input_shape = (128, 128, 3)
num_classes = 2
image_resize = 128
data_generator = ImageDataGenerator(
    preprocessing_function=preprocess_input,
)
```

```python
test_generator = data_generator.flow_from_directory(
    'E://test1W',
    target_size=(image_resize, image_resize),
    shuffle=False)
```

```python
googleNet_model = InceptionResNetV2(include_top=False, weights='imagenet', input_shape=input_shape)
googleNet_model.trainable = True
model = Sequential()
model.add(googleNet_model)
model.add(GlobalAveragePooling2D())
model.add(Dense(units=2, activation='softmax'))
model.compile(loss='binary_crossentropy',
              optimizer=optimizers.Adam(lr=1e-5, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False),
              metrics=['accuracy'])
model.summary()
```

```python
from tensorflow.keras.models import load_model  
model.load_weights('deepfake-detection-model_weight.h5')
```

```python
model.summary()
```

```python
model.evaluate_generator(test_generator)
```

```python
prediction = model.predict_generator(test_generator)
```

```python
import re
```

```python
def original(name):
    name=re.sub('\d+', '', name)
    name=name.split(".")[0]+".mp4"
    return name
```

```python
df['original']=df['pictures'].apply(original)
```

```python
df.head(5)
```

```python
original=[]
for i in df["original"]:
    if i in original:
        next
    else:
        original.append(i)
```

```python
pre = []
for i in original:
    print(i)
    video=df[df['original']==i]
    Fake_count=len(video[video['pre_label']==0])/len(video)
    Real_count=len(video[video['pre_label']==1])/len(video)
    if (Fake_count>0.5) or (Real_count<0.5):
        pre.append(0)
    else:
        pre.append(1)
```
